```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('datasets\Cost_of_Living_Index_by_Country.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df_sorted = df.sort_values(by='Cost of Living Index', ascending=False)
```

```python
top_10_highest = df_sorted.head(10)
top_10_highest[['Country','Cost of Living Index']]
```

```python
top_10_lowest = df_sorted.tail(10)
top_10_lowest[['Country','Cost of Living Index']]
```

```python
plt.figure(figsize=(14, 8))

plt.subplot(1,2,1)
sns.barplot(data=top_10_highest, x='Cost of Living Index', y='Country', palette='inferno')
plt.title('Top 10 Most Expensive Countries')
plt.xlabel('Cost of Living Index')

plt.subplot(1,2,2)
sns.barplot(data=top_10_lowest, x='Cost of Living Index', y='Country', palette='plasma')
plt.title('Top 10 Most Cheapest Countries')
plt.xlabel('Cost of Living Index')

plt.tight_layout()
plt.show()
```

```python
df.head(2)
```

```python
columns = ['Rent Index','Groceries Index','Restaurant Price Index', 'Local Purchasing Power Index']
for column in columns:
    sns.barplot(data=top_10_highest, x=column, y='Country', palette='inferno')
    plt.title(f'{column} for top 10 Expensive country')
    plt.show()
```

```python
for column in columns:
    sns.barplot(data=top_10_lowest, x=column, y='Country', palette='inferno')
    plt.title(f'{column} for top 10 cheapest country')
    plt.show()
```

```python
def cost_of_living_class(index):
    if index > 80:
        return 'high'
    elif index > 50:
        return 'medium'
    else:
        return 'low'
```

```python
df ['cost of living class'] = df['Cost of Living Index'].apply(cost_of_living_class)
df.head(2)
```
